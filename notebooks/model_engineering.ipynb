```python
import pandas as pd 
from sklearn.preprocessing import StandardScaler
```

```python
#arquivo limpO
caminho_dados = '../data/train_clean.csv' 
df = pd.read_csv(caminho_dados)

# printa dimensão e as primeiras linhas
print("Dimensão do dataset:", df.shape)
df.head()
```

```python
#O target é a coluna winner, que indica o vencedor de cada jogo
y = df['winner']

# Retiramos a winner do dataset para criar o x 
X = df.drop(columns=['winner'])

print("Formato de X:", X.shape, "| Formato de y:", y.shape)
X.head()
```

```python

#binarizando o nome dos paises usando onehot encoding e evitando redundancias com o drop_first=true
X_encoded = pd.get_dummies(X, drop_first=True)
colunas_x = X_encoded.columns

print("Novas colunas em X_encoded:", X_encoded.shape[1])
X_encoded.head()
```

```python

scaler = StandardScaler()
colunas_x = X_encoded.columns
X_escalonado = scaler.fit_transform(X_encoded)  
X_final = pd.DataFrame(X_escalonado, columns=colunas_x)

print("escalonamento concluído. primeiras linhas de X_final:")  
X_final.head()
```

```python
# Juntamos tudo para exportar num único arquivo
dataset_pronto = X_final.copy()
dataset_pronto['winner'] = y.values 

# Salva o arquivo
dataset_pronto.to_csv('../data/train_engineered.csv', index=False)
print("'train_engineered.csv' salvo com sucesso!")
```

```python
#APLICANDO A MESMA LOGICA DAS CELULAS ANTERIORES NO ARQUIVO DE TESTE



# Carregar o arquivo de teste limpo
df_test = pd.read_csv('../data/test_clean.csv')


print("Dimensão do dataset de teste:", df_test.shape)

# Separar X e y caso o arquivo de teste possua a coluna 'winner'
if 'winner' in df_test.columns:
    X_test_raw = df_test.drop(columns=['winner'])
    y_test = df_test['winner']
else:
    X_test_raw = df_test
    y_test = None

# Transformação Categórica
X_test_encoded = pd.get_dummies(X_test_raw, drop_first=True)

# Garante que o teste tenha as mesmas colunas que o treino

X_test_encoded = X_test_encoded.reindex(columns=colunas_x, fill_value=0)

# Escalonamento
X_test_escalonado = scaler.transform(X_test_encoded)

# Transformamos de volta para um DataFrame 
X_test_final = pd.DataFrame(X_test_escalonado, columns=colunas_x)

# Salvar o arquivo final
dataset_teste_pronto = X_test_final.copy()
if y_test is not None:
    dataset_teste_pronto['winner'] = y_test.values 

# Salvamos na pasta data com um novo nome
dataset_teste_pronto.to_csv('../data/test_engineered.csv', index=False)
print("Arquivo 'test_engineered.csv' processado e salvo com sucesso!")
```
